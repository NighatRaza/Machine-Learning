# TASK 1: IRIS FLOWER CLASSIFICATION

# importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# loading the dataset

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
column_names = ['id','sepal_length','sepal_width','petal_length','petal_width','species']
df.columns = column_names
df.head()
```

```python
df.describe()
```

# visualization

```python
sns.pairplot(df,hue='species')
```

# separating input features and output outcome

```python
X = df[['sepal_length','sepal_width','petal_length','petal_width']].values
Y = df['species'].values
```

# splitting into training and testing data

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

## train model using support vector machine algorithm 

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train,y_train)
```

```python
svc_prediction = model_svc.predict(X_test)
```

### calculating accuracy score

```python
print(accuracy_score(y_test, svc_prediction))
```

## train model using logistic regression

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(X_train,y_train)
```

```python
LR_prediction = model_LR.predict(X_test)
```

### calculating accuracy score

```python
print(accuracy_score(y_test, LR_prediction))
```

## train model using decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
model_DTC = DecisionTreeClassifier()
model_DTC.fit(X_train,y_train)
```

```python
DTC_prediction = model_DTC.predict(X_test)
```

### calculating accuracy score

```python
print(accuracy_score(y_test, DTC_prediction))
```

# classification report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,LR_prediction))
```
