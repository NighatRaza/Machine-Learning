# binary classification - supervised

```python
# a submarine send sonar signals which detect obstacles in sea i.e mine and rock
# a ml model to predict whether obstacle is mine or a rock
```

# importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# data collection

```python
df = pd.read_csv('sonar data.csv',header=None)
df.head()
```

# data pre processing

```python
df.shape
```

```python
# DATA
# ----------

# COLUMNS:
# total features = 60
# target = 1 (R: Rock, M:Mine)

# ROWS:
# 208 records
```

```python
df.describe()
```

```python
df[60].value_counts()
```

```python
df.groupby(60).mean()
```

# dividing features and target

```python
X = df.drop(columns=60,axis=1)
y = df[60]
print(X)
print(y)
```

# train test split

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,stratify=y,random_state=42)
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

```python
print(y.shape, y_train.shape, y_test.shape)
```

```python
print(x_train,y_train)
```

```python
print(x_test,y_test)
```

# logistic regression

```python
model = LogisticRegression()
```

# train

```python
model.fit(x_train,y_train)
```

# model evaluation

```python
# accuracy on training data
```

```python
train_predict = model.predict(x_train)
train_data_accuracy = accuracy_score(train_predict, y_train)
train_data_accuracy
```

```python
# accuracy on testing data
```

```python
# unseen data
test_predict = model.predict(x_test)
test_data_accuracy = accuracy_score(test_predict, y_test)
test_data_accuracy
```
